```python
from transformers import pipeline

# Load the pre-trained models
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
qa_model = pipeline("question-answering")
```

```python
# Function for text summarization
def generate_summary(text):
    summary = summarizer(text, max_length=150, min_length=40, do_sample=False)
    return summary[0]['summary_text']

# Function for question answering
def generate_qa(context, question):
    return qa_model(question=question, context=context)
```

```python
# Sample text (could be part of a syllabus, study material, etc.)
text = """
Artificial Intelligence (AI) is intelligence demonstrated by machines, in contrast to the natural intelligence displayed by humans and animals. Leading AI textbooks define the field as the study of "intelligent agents": any device that perceives its environment and takes actions that maximize its chance of successfully achieving its goals. Colloquially, the term "artificial intelligence" is often used to describe machines (or computers) that mimic "cognitive" functions that humans associate with the human mind, such as "learning" and "problem-solving".
"""

# Test the summarization function
summary = generate_summary(text)
print("Summary:")
print(summary)

# Test the question-answering function
question = "What is Artificial Intelligence?"
answer = generate_qa(text, question)
print("\nAnswer to the question:")
print(answer['answer'])
```

```python
# Function to generate flashcards from the text
def generate_flashcards(text):
    # Split the text into sentences (basic approach)
    sentences = text.split('.')[:5]  # Limit to the first 5 sentences
    flashcards = []

    # Create flashcards from each sentence
    for sentence in sentences:
        if len(sentence.strip()) > 10:  # Ignore very short sentences
            flashcards.append({
                "Q": f"What is: {sentence.strip()}?",
                "A": sentence.strip()
            })

    return flashcards

# Test the flashcard generation function
flashcards = generate_flashcards(text)

# Print the generated flashcards
print("Generated Flashcards:")
for card in flashcards:
    print(f"Q: {card['Q']}\nA: {card['A']}\n")
```

```python
%%writefile app.py
import streamlit as st
from transformers import pipeline

# Load the models
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
qa_model = pipeline("question-answering")

# Functions for summarization, question answering, and flashcard generation
def generate_summary(text):
    summary = summarizer(text, max_length=150, min_length=40, do_sample=False)
    return summary[0]['summary_text']

def generate_qa(context, question):
    return qa_model(question=question, context=context)

def generate_flashcards(text):
    sentences = text.split('.')[:5]
    flashcards = []
    for sentence in sentences:
        if len(sentence.strip()) > 10:
            flashcards.append({"Q": f"What is: {sentence.strip()}?", "A": sentence.strip()})
    return flashcards

# Streamlit UI
st.title("AI StudyBuddy")

# Input section for text (syllabus or study material)
text = st.text_area("Enter your text here:", height=200)

# Button to generate summary
if st.button("Generate Summary"):
    if text:
        summary = generate_summary(text)
        st.write("Summary:")
        st.write(summary)
    else:
        st.warning("Please enter some text to summarize.")

# Input section for question answering
question = st.text_input("Ask a question about the text:")
if question:
    if text:
        answer = generate_qa(text, question)
        st.write("Answer:")
        st.write(answer['answer'])
    else:
        st.warning("Please enter the text first.")

# Button to generate flashcards
if st.button("Generate Flashcards"):
    if text:
        flashcards = generate_flashcards(text)
        st.write("Generated Flashcards:")
        for card in flashcards:
            st.write(f"Q: {card['Q']}\nA: {card['A']}\n")
    else:
        st.warning("Please enter some text to generate flashcards.")
```

```python
%%writefile app.py
import streamlit as st
from transformers import pipeline
from PIL import Image

# Load models
summarizer = pipeline("summarization")
question_answerer = pipeline("question-answering")

# Sidebar Navigation
st.sidebar.title("AI StudyBuddy 🚀")
page = st.sidebar.radio("Go to", ["📄 Summarizer", "❓ Q&A", "🧠 Flashcards"])

st.title("📚 AI StudyBuddy")

if page == "📄 Summarizer":
    st.subheader("Text Summarization")
    text = st.text_area("Enter text to summarize", height=300)
    if st.button("Summarize"):
        with st.spinner("Summarizing..."):
            summary = summarizer(text, max_length=150, min_length=40, do_sample=False)
            st.success(summary[0]['summary_text'])

elif page == "❓ Q&A":
    st.subheader("Ask a Question")
    context = st.text_area("Enter context passage", height=200)
    question = st.text_input("Enter your question")
    if st.button("Get Answer"):
        with st.spinner("Thinking..."):
            answer = question_answerer(question=question, context=context)
            st.success(answer['answer'])

elif page == "🧠 Flashcards":
    st.subheader("Flashcard Generator (coming soon!)")
    st.info("This feature will generate flashcards from your notes.")
```
